```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("shopping_trends.csv")
df
```

```python
# histogram plot of age vs frequency
df['Age'].plot(kind = "hist", edgecolor = "black")
plt.xlabel("Age")
plt.title("Age vs Frequency")
plt.show()
```

```python
# bar plot of gender vs count
df["Gender"].value_counts().plot(kind = 'bar', color = ['blue', 'orange'])
plt.title("Gender vs count")
plt.grid(alpha = 0.8)
plt.show()
```

### Used to highlight discrete categories, shows gender counts over a dataset, and also shows change over time


```python
# Average of Purchase amount by category
df.groupby("Category")['Purchase Amount (USD)'].mean().sort_values(ascending = False).tail(3)
```

```python
# Review Rating in Descending order
df.sort_values("Review Rating", ascending = False, inplace = True)
df
```

```python
# Count of Subscription Status
df['Subscription Status'].value_counts().plot(kind = 'pie', autopct = '%1.1f%%' )
plt.title("Pie chart of Subscription Status Count")
plt.show()
```

### This pie chart shows data slices of a circle, each slice representing the percentage of each categries 
### this pie chart shows of Subscription Status count in Yes and No format

```python
# plotting a box plot for Purchased Amount by Frequency of Purchases
plt.figure(figsize= (10, 10))
df.boxplot(column ='Purchase Amount (USD)',by = "Frequency of Purchases")
plt.title("")
plt.show()
```

### Box plot shows a Outlier in data on frequency of perchases, on Annually ,monthly and weekly 

```python
# Count of Payment Method
# Count the values
counts = df['Payment Method'].value_counts()

# Generate distinct colors
num_bars = len(counts)
colors = plt.cm.tab20(np.linspace(0, 1, num_bars))

# Create a plot and save the axes object
ax = counts.plot(kind='bar', edgecolor='black', color=colors)

# Add title and labels (optional)
plt.ylabel('Count')
plt.title('---')

# Annotate each bar with its value
for i, value in enumerate(counts):
    ax.text(i, value + 1, str(value), ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()
```

```python
# pivot table for average of Purchased Amount according to its location and item purchased
pd.pivot_table(df, values = 'Purchase Amount (USD)', index = "Location", columns = "Item Purchased", aggfunc = "mean").head(10)
```

```python
df.groupby("Payment Method")["Purchase Amount (USD)"].sum().sort_values(ascending = False).plot(kind = "line")
plt.title("Sum of Purchased Method in descending order based on Payement type ")
plt.show()
```

### Sum of purchaed method in descending order based on payment type
### Using line chart to shows collection of point to line, perfect show trends and patterns along going axis like time.

```python
# Median of Previous Purchases groupedby Frequency of Purchases
df.groupby("Frequency of Purchases")['Previous Purchases'].median()
```

```python
# Count of Season
df['Season'].value_counts().plot(kind = 'bar')
plt.title("Season Count")
plt.grid()
plt.show()
```
